```python
import matplotlib.pyplot as plt

# For Jupter Notebook
%matplotlib inline

import numpy as np
```

```python
x = np.linspace(-1.2,1.2,100)
x
```

```python

fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(15,5))

for i in range(1,5):
    ax.plot(x,x**i,label = i) 
    
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.legend()
ax.set_title("Plot")
```

```python
#Implicit
for i in range(1,5):
    plt.plot(x,x**i,label = i) 
    
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Plot")
plt.legend()
```

```python
df1 = df.construction.value_counts()
df1.plot(kind="pie")
```

```python
fig,ax = plt.subplots()
ax.pie(df1,labels = df1.index,autopct="%1.2f%%",explode=(0.01,0.1,0.5,0.5,0.5));
ax.set_title("Construction")
```
